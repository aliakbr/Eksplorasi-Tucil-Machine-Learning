### Poin a Pembacaan Dataset

Pada poin ini kami akan membaca dataset menggunakan modul pandas read_csv. Agar read_csv tidak menghasilkan error, file 'iris.csv' dan 'playtennis.csv' harus berada pada direktori yang sama

Berikut adalah data dari Iris CSV

```python
import pandas as pd
import numpy as np
pd1 = pd.read_csv('iris.csv')
print(pd1)
```

Dan berikut adalah data dari playtennis

```python
pd2 = pd.read_csv('playtennis.csv')
print(pd2)
```

```python
Xiris = pd1.iloc[:,0:4].values.tolist()
print(Xiris)
Yiris = pd1.iloc[:,-1].values.tolist()
print(Yiris)
```

## Poin b : Melakukan Pembelajaran dengan DTL dan MLP

```python
# Full-training DTL
from sklearn import tree

clf1 = tree.DecisionTreeClassifier()
clf1 = clf1.fit(Xiris,Yiris)
```

```python
# Visualisasi DTL
import graphviz

dot_data = tree.export_graphviz(clf1, out_file=None)
graph = graphviz.Source(dot_data)
graph
```

```python
#Full-training MLP
from sklearn.neural_network import MLPClassifier
clf2 = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(15, 3), random_state=1)
clf2.fit(Xiris, Yiris)
```

```python
#Visualisasi MLP
import matplotlib.pyplot as plt

fig, axes = plt.subplots(4,4)
vmin, vmax = clf2.coefs_[0].min(), clf2.coefs_[0].max()
for coef, ax in zip(clf2.coefs_[0].T, axes.ravel()):
    ax.matshow(coef.reshape(1,4), cmap=plt.cm.gray, vmin=.5 * vmin, vmax=.5 * vmax)
    ax.set_xticks(())
    ax.set_yticks(())
    
plt.show()
```

## Poin c : Pembelajaran dengan Split Test, 90% untuk Train dan 10% untuk tes

```python
#Split data 90%/10%
from sklearn.model_selection import train_test_split, KFold, cross_val_score

X_train, X_test, Y_train, Y_test = train_test_split(Xiris, Yiris, test_size=0.1, random_state=42)
print(X_train)
print(X_test)
```

```python
#Split data training DTL
clf3 = tree.DecisionTreeClassifier()
clf3 = clf3.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
# Kinerja DTL
# Akurasi
y_pred = clf3.predict(X_test)
print('Akukrasi =', accuracy_score(Y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(Y_test, y_pred)
print('Confusion Matrix =\n', cm)
```

```python
#Full-training MLP
clf4 = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(15, 3), random_state=1)
clf4.fit(X_train, Y_train)
```

```python
# Kinerja MLP
# Akurasi
y_pred = clf4.predict(X_test)
print('Akukrasi =', accuracy_score(Y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(Y_test, y_pred)
print('Confusion Matrix =\n', cm)
```

## Poin d : Pembelajaran untuk 10-Fold Cross Validation

```python
#Cross validation training DTL
clf5 = tree.DecisionTreeClassifier()
scores = cross_val_score(clf5, Xiris, Yiris, cv=10)
print('Akurasi/fold =', scores)
print("Akurasi: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
#Cross validation training MLP
clf6 = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(15, 3), random_state=1)
scores = cross_val_score(clf6, Xiris, Yiris, cv=10)
print('Akurasi/fold =', scores)
print("Akurasi: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

Apabila antara garis aproksimasi regresi menunjukkan adanya kemonotonan (seperti sepallength dan sepalwidth pada kelas Iris-Setosa) maka kedua feature berkorelasi terhadap kelas. Sementara jika tidak ada kemonotonan (seperti sepalwidth dan petalwidth pada kelas Iris-Setosa) maka kedua feature tidak berkorelasi pada kelas tersebut.
